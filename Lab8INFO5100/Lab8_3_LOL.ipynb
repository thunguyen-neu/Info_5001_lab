```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('LOL/140325_LoL_champion_data.csv')

heatmap_data = df.pivot_table(index='herotype', columns='rangetype', aggfunc='size', fill_value=0)
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm')

plt.title('Heatmap of Champion Types')
plt.xlabel('Range Type')
plt.ylabel('Hero Type')
plt.tight_layout()
plt.show()
```
