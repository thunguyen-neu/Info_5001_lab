```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Global_AI/Global_AI_Content_Impact_Dataset.csv')
industry_year_counts = df.groupby(['Industry', 'Year']).size().unstack(fill_value=0)
plt.figure(figsize=(14, 7))
industry_year_counts.T.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('AI Content Impact: Number of Entries per Industry by Year')
plt.xlabel('Year')
plt.ylabel('Number of Entries')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
