```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon/all_pokemon_data.csv')
type_counts = df['Primary Typing'].value_counts()
plt.figure(figsize=(12, 6))
type_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Pokémon by Primary Type')
plt.xlabel('Primary Typing')
plt.ylabel('Number of Pokémon')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
