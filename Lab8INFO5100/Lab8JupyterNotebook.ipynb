## Line Plot

### Univariate
Visual representation of the relationship of a single column in a dataset.

````python
plt.grid(True)

plt.plot(iris_dataframe["Id"], iris_dataframe['SepalWidthCm'])
plt.title("Iris Sepal Length")
plt.xlabel("Id")
plt.ylabel("Sepal Width (cm)")
plt.xticks(range(0, max(iris_dataframe["Id"]) +1,10))
plt.show()
````

````python
# adding gird-lines
plt.grid(True)

plt.plot(iris_dataframe["Id"], iris_dataframe['SepalLengthCm'])
plt.title("Iris Sepal Length")
plt.xlabel("Id")
plt.ylabel("Sepal Length (cm)")
# Set x-axis for every 10 units
plt.xticks(range(0, max(iris_dataframe["Id"]) + 1, 10))  
plt.show()
````

## Scatter Plot

### Bivariate
Visual representation of the relationship between two variables in a dataset.

````python
plt.grid(True)
plt.scatter(iris_dataframe["PetalLengthCm"], iris_dataframe["PetalWidthCm"])
plt.title("Iris Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
````

### Multivariate
Visual representation of the relationship between three or more columns/ variables in a dataset.

````python
# Species
# Iris-setosa        50
# Iris-versicolor    50
# Iris-virginica     50

iris_species_color = {"Iris-setosa": "red", 
                      "Iris-versicolor": "green", 
                      "Iris-virginica":"blue"}

for species, species_color in iris_species_color.items():
    # creating a data frame for a specific value for 'Species' column:
    species_dataframe = iris_dataframe[iris_dataframe["Species"] == species]
    # parameter c - color
    plt.scatter(species_dataframe["PetalLengthCm"], species_dataframe["PetalWidthCm"], c = species_color, label=species)

plt.grid(True)
plt.title("Iris Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
````

## Histograms

````python
sns.histplot(data=iris_dataframe, x="SepalLengthCm")
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
````

````python
sns.histplot(data=iris_dataframe, x="SepalLengthCm")
plt.title("Sepal Length Distribution")
plt.grid(axis="y")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.yticks(range(0, 30, 2))
````

## Pair Plot
Introduce pairwise relationships.

````python
sns.pairplot(iris_dataframe)
plt.show()
````

## Box Plots

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_dataframe, palette="Set3")

# Add labels and title
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
````

## Swarm Plot + Box Plot

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_dataframe, palette="pastel")
sns.swarmplot(x="Species", y="SepalLengthCm", data=iris_dataframe, color=".25")
plt.title("Box and Swarm Plot of Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
````

## Kernel Density Plot (KDE)

KDE plot helps to visualize the distribution of data in a dataset



````python
# Create a KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=iris_dataframe, x="SepalLengthCm", hue="Species", fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Kernel Density Plot of Sepal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Density")
plt.show()
````

## Violin Plot (Box Plot + KDE)

A statistical graphic that shows the distribution of numerical data by combining a box plot and a kernel density plot.

````python
sns.violinplot(x='Species', y='PetalWidthCm', data=iris_dataframe)
plt.title('Violin Plot of Petal Width by Species')
plt.show()
````

## Correlation Heat Maps

````python
correlation = iris_dataframe.iloc[:, :-1].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correl ation Heatmap')
plt.show()
````
